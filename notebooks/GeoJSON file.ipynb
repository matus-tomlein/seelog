```python
import shapefile
import pandas as pd
import numpy as np
from shapely.geometry import Polygon, MultiPolygon, Point
import json
import geohash
```

```python
countries = json.loads(open("ne_10m_admin_0_countries/countries.geojson").read())
```

```python
features = countries['features']
```

```python
# geohash.decode('hfads')
geohash.encode(-52.9, 180)
```

```python
all_geohashes = {}
to_process = ['pbpb']

while len(to_process) > 0:
    gh = to_process.pop()
    all_geohashes[gh] = True
    
    for neighbor in geohash.neighbors(gh):
        if neighbor not in all_geohashes:
            to_process.append(neighbor)
```

```python
country_polygons = {}

for feature in features:
    coordinates = feature['geometry']['coordinates']
    while type(coordinates[0][0][0]) == list:
        coordinates = [n for i in coordinates for n in i]
        
    country_polygons[(feature['properties']['ADMIN'], feature['properties']['ISO_A3'])] = [Polygon(c) for c in coordinates]
```

```python
def country_distance_to_pos(country, pos):
    return min([coordinates.distance(Point(pos[0], pos[1])) for coordinates in country_polygons[country]])
```

```python
def get_closest_country(pos):
    distances = [(country, country_distance_to_pos(country, pos)) for country in country_polygons]
    closest_country = min(distances, key=lambda x: x[1])[0]
    return closest_country
```

```python
countries = [key for key in country_polygons]
```

```python
for gh in all_geohashes:
    pos = geohash.decode(gh)
    all_geohashes[gh] = countries.index(get_closest_country(pos))
```
